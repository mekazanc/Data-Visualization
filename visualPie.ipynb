Pie Chart 

In this tutorial, I applied some visualization techniques that is so simple. You can simply use in your work.

I took them from a very nice tutorial whose link is below.

https://github.com/CoreyMSchafer/code_snippets/tree/master/Python/Matplotlib/03-PieCharts



```
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
```

```
# Give data and labels 
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# Emphasize some parts 
explode = [0, 0, 0, 0.1, 0]


# Plot pie chart
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

 
# Add title
plt.title("My First Pie Chart")
plt.tight_layout()
plt.show()
```
