In addition to that you can fill difference between two plots like below. 

```
plt.plot(ages, java, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, python, label='Python')



overall_median = 50000
# can use where=(python <= overall_median) parameter as well. 

plt.fill_between(ages, python, java,
                 where=(python > java),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, python, java,
                 where=(python <= java),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')



plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()



```
